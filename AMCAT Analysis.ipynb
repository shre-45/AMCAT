#### CATEGORICAL AND CATEGORICAL (RESEARCH QUESTION)

```python
# Assuming 'Designation' and 'Gender' are columns in your DataFrame df
top_designations = df['Designation'].value_counts().head(10)  # Adjust the number to select top N designations
df_top_designations = df[df['Designation'].isin(top_designations.index)]
gender_distribution = df_top_designations.groupby('Designation')['Gender'].value_counts().unstack().fillna(0)
gender_distribution.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Gender Distribution within Top Designations')
plt.xlabel('Designation')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.show()
```

* The above stacked bar plot tells gives insights about designation according to gender.
* The blue are female and orange are male.
* The male and female both the genders of top designation is software engineers.
* There is a relationship with the gender for the designation comparing with gender male are more with the software designations.
